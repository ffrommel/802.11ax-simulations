```python
# Constantes 
l = 9.8 # Largo de la sala (m)
a = 7.4 # Ancho de la sala (m)
```

```python
from math import sqrt
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
# Calcular separaci√≥n en largo y ancho de STAs
def separacion_l_a(stas):
    lado = sqrt((l*a) / stas) # lado del cuadrado de cada STA
    stas_l = round(l / lado)  # #STAs en largo
    stas_a = round(a / lado)  # #STAs en ancho 
    if stas_l * stas_a < stas:
        stas_a = stas_a + 1
    return l / stas_l, a / stas_a
```

```python
# Obtener coordenadas
def coordenadas(stas):
    sep_l, sep_a = separacion_l_a(stas)
    print("sep largo: ", round(sep_l, 2), " - sep ancho: ", round(sep_a, 2))
    print("")
    i = sep_l / 2
    j = sep_a / 2
    cont = 1
    vect_x = []
    vect_y = []
    #print("Coordenadas:")
    while i < l:
        while j <= a:
            if cont <= stas:
                print("(" + str(round(i, 2)) + ",\t" + str(round(j, 2)) + ")")
                #print("(" + str(i) + ",\t" + str(j) + ")")
                #print("")
                vect_x.append(i)
                vect_y.append(j)
            j = j + sep_a
            cont = cont + 1
        i = i + sep_l
        j = sep_a / 2
    return vect_x, vect_y
```

```python
# Graficar 1 grilla
sta = 12
vect_x, vect_y = coordenadas(sta)
plt.plot(vect_x, vect_y, 'o', color='black') # STAs
plt.plot(l / 2, a / 2, 'x', color='red') # AP
plt.xlim(0, l)
plt.ylim(0, a)
plt.title("#STAs " + str(sta))
```

```python
# Graficar todas las grillas
vect_stas = [1] + [i for i in range(3, 54, 3)]
#vect_stas = [6, 12, 18, 24]
#print("#STAs a considerar:")
#print(vect_stas)
#print("")
vect_stas_it = iter(vect_stas)
cols = 4
rows = round(len(vect_stas) / cols) + 1
fig, axs = plt.subplots(rows, cols, figsize=(15, 15))
for i in range(0, rows):
    for j in range(0, cols):
        try:
            sta = next(vect_stas_it)
            print("")
            print("#STAs: ", sta)
        except:
            break
        vect_x, vect_y = coordenadas(sta)
        axs[i, j].plot(vect_x, vect_y, 'o', color='black') # STAs
        axs[i, j].plot(l / 2, a / 2, 'x', color='red') # AP
        axs[i, j].set_xlim(0, l)
        axs[i, j].set_ylim(0, a)
        axs[i, j].set_title("#STAs " + str(sta))
```
