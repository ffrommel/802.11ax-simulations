```python
# Constantes 
l = 9.8 # Largo de la sala (m)
a = 7.4 # Ancho de la sala (m)
```

```python
from math import sqrt
```

```python
# Calcular separaci√≥n en largo y ancho de STAs
def separacion_l_a(stas):
    lado = sqrt((l*a) / stas) # lado del cuadrado de cada STA
    stas_l = round(l / lado)  # #STAs en largo
    stas_a = round(a / lado)  # #STAs en ancho 
    if stas_l * stas_a < stas:
        stas_a = stas_a + 1
    return l / stas_l, a / stas_a
```

```python
# Obtener coordenadas
def coordenadas(stas):
    sep_l, sep_a = separacion_l_a(stas)
    print("sep largo: ", round(sep_l, 2), " - sep ancho: ", round(sep_a, 2))
    print("")
    i = sep_l / 2
    j = sep_a / 2
    cont = 1
    vect_x = []
    vect_y = []
    #print("Coordenadas:")
    while i < l:
        while j <= a:
            if cont <= stas:
                print("(" + str(round(i, 2)) + ",\t" + str(round(j, 2)) + ")")
                #print("")
                vect_x.append(i)
                vect_y.append(j)
            j = j + sep_a
            cont = cont + 1
        i = i + sep_l
        j = sep_a / 2
    return vect_x, vect_y
```

```python
# Generar todos los archivos de configuraci√≥n de Komondor
vect_stas = [1] + [i for i in range(3, 54, 3)]
for sta in vect_stas:
    vect_x, vect_y = coordenadas(sta)
    with open('input/input_' + str(sta) + 'sta.csv','w') as file:
        # Escribir header
        file.write("node_code;node_type;wlan_code;x(m);y(m);z(m);central_freq (GHz);channel_bonding_model;primary_channel;min_channel_allowed;max_channel_allowed;tx_power;sensitivity;traffic_model;traffic_load(pkts/s);packet_length;num_packets_aggregated;capture_effect_thr;capture_effect_thr;constant PER;pifs_activated;cw_adaptation;cw;cw_stage\n")
        # Escribir AP
        file.write("AP;0;A;" + str(l / 2) + ";" + str(a / 2) + ";2;5;4;0;0;1;15;-82;99;1000;12000;64;0;10;0;0;0;16;5\n")
        # Escribir STAs
        for i in range(0, sta):
            file.write("STA" + str(i + 1) + ";1;A;" + str(vect_x[i]) + ";" + str(vect_y[i]) + ";0;5;4;0;0;1;15;-82;99;1000;12000;64;0;10;0;0;0;16;5\n")
```
